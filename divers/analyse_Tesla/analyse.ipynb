Analyse des données Tesla

```python
import pandas as pd
```

```python
q2 = pd.read_csv('FR-Q2.csv')
q3 = pd.read_csv('FR-Q3.csv')
q23 = pd.concat([q2, q3])
```

```python
#q23 = q23.rename(columns={'id_pdc_iterance':'id_pdc_itinerance'})
q23['month'] = pd.to_datetime(q23['start']).dt.month
```

```python
pdc_stat = pd.read_csv('lien_station_pdc.csv')
q23_pdc = pd.merge(
    q23,
    pdc_stat,
    how="left",
    on='id_pdc_itinerance'
)
q23_pdc
```

```python
q23_stat = q23_pdc[['energy', 'id_station_itinerance', 'month']].groupby(["id_station_itinerance", "month"]).sum().reset_index()
q23_stat
```

```python
q23_stat.to_csv('q23_stat.csv')
```
