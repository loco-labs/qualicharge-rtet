# Validation des numÃ©ros SIREN

```python
def valid_siren(siren: str) -> bool:
    if len(siren) != 9 :
        return False
    numero = [int(char) for char in siren]
    produit = [ num * coef for num, coef in zip(numero, [1,2,1,2,1,2,1,2,1])]
    somme = [ prod % 10 + prod // 10 for prod in produit]
    resultat = sum(somme) % 10 == 0 and sum(somme) > 1
    return resultat
```

```python
tests = ( ['55210055', False], ['552100554', True], ['123456789', False], ['000000000', False], ['948330824', True], ['948331824', False], ['665580148', True], ['765580148', False])
```

```python
for test in tests: 
    assert valid_siren(test[0]) == test[1]
```
