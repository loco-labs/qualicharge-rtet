```python
import pandas as pd
```

```python
df =pd.DataFrame({
    'entity': ['am1', 'am1', 'am1', 'am1', 'am2', 'am2', 'am2', 'am2'],
    'siren': [1,1,1,1,2,2,2,2],
    'code': ['FR1', 'FR1', 'FR2', 'FR2', 'FR3', 'FR3', 'FR3', 'FR4'],
    'from': ['1900-01-01'] * 8,
    'to': ['1900-01-02'] * 8,
    'id': ['st1','st2','st3','st4','st5','st6','st7','st8'],
    'energy': [10, 20, 30, 40, 50, 60, 70, 80]})
df
```

```python
df_station = df.groupby(['entity', 'siren', 'code', 'from', 'to'])[['id', 'energy']].apply(lambda x: x.to_dict(orient='records')).rename("stations").reset_index()
df_unit = df_station.groupby(['entity', 'siren'])[['code', 'from', 'to', 'stations']].apply(lambda x: x.to_dict(orient='records')).rename("units").reset_index()
df_unit.to_dict(orient='records')
```
