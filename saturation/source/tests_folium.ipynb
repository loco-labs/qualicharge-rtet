```python

points = [
    {
        "time": "2017-06-02",
        "color": "green",
        "coordinates": [-2.548828, 51.467697],
    },
        {
        "time": "2017-07-02",
        "color": "green",
        "coordinates": [-2.548828, 51.467697],
    },
    {
        "time": "2017-08-02",
        "color": "red",
        "coordinates": [-2.548828, 51.467697],
    },
    {
        "time": "2017-09-02",
        "color": "green",
        "coordinates": [-2.548828, 51.467697],
    },
    {
        "time": "2017-10-02",
        "color": "green",
        "coordinates": [-2.548828, 51.467697],
    },
    {
        "time": "2017-07-02",
        "color": "blue",
        "coordinates": [-0.087891, 51.536086],
    },
    {
        "time": "2017-08-02",
        "color": "purple",
        "coordinates": [-6.240234, 53.383328],
    },
    {
        "time": "2017-09-02",
        "color": "black",
        "coordinates": [-1.40625, 60.261617],
    },
    {
        "time": "2017-10-02", 
        "color": "yellow",
        "coordinates": [-1.516113, 53.800651]},
]

features = [
    {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": point["coordinates"],
        },
        "properties": {
            "time": point["time"],
            "icon": "circle",
            "style": {
                "color": point["color"],
                "weight": 0,
                "radius": 5,
                "fillOpacity": 1
            },     
        },
    }
    for point in points
]

m = folium.Map(
    location=[56.096555, -3.64746],
    tiles="cartodbpositron",
    zoom_start=5,
)
#m = carte
folium.plugins.TimestampedGeoJson(
    {"type": "FeatureCollection", "features": features},
    period="P1M",
    auto_play=False,
    loop=False,
    max_speed=1,
    loop_button=True,
    date_options="YYYY/MM/DD",
    time_slider_drag_update=True,
    duration="P0M",
).add_to(m)

m
```
